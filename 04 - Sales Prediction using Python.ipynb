## 5 - Data Preprocessing

### 5.1 - Data Cleaning :

```python
data.drop_duplicates(inplace=True)
```

```python
# Removing Unecessary columns like Customer name, customer e-mail
data = data.drop(columns =['customer name','customer e-mail'], axis = 1 )
```

```python
data.head()
```

### 5.2 - Label Encoding :

```python
# label encoding on country column to replace text values with numerical values
label = LabelEncoder()
```

```python
data['country'] = label.fit_transform(data['country'])
```

### 5.3 - Separating Feature & Target Variables

```python
x = data.drop(columns = ['car purchase amount'], axis =1)
y = data['car purchase amount']
```

### 5.4 - Data Standardization :

```python
scaler = StandardScaler()
```

### 5.5 - Spliting Data into Training & Testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

## 6 - Model Building & Training

For this Regression task like predicting a continuos value, we'll use Random Forest Regresssion or Linear Regression model.

```python
regression = RandomForestRegressor(n_estimators=100)
```

```python
regression.fit(x_train, y_train)
```

```python
#using linear regression
model = LinearRegression()
model.fit(x_train,y_train)
```

## 7 - Model Evaluation

```python
#Accuracy of training data using Random forest regression
x_train_prediction = regression.predict(x_train)
mae = mean_absolute_error(x_train_prediction, y_train)
print("Mean absolute error: ", mae)
```

```python
#Accuracy of training data using Linear Regression
x_train1_prediction = model.predict(x_train)
mae1 = mean_absolute_error(x_train1_prediction, y_train)
print("Mean absolute error: ", mae1)
```

## 8 - Model Deployement

```python
# Create a dictionary to store user input
input_data = {}

# Define the list of features (replace with your actual feature names)
features = ['country', 'gender', 'age', 'annual Salary', 'credit card debt', 'net worth']

# Loop through each feature to get user input
for feature in features:
    if feature == 'country':
        value = input(f"Enter {feature}: ")
    else:
        while True:
            try:
                value = float(input(f"Enter {feature}: "))
                break  # Exit the loop if input is valid
            except ValueError:
                print("Invalid input. Please enter a numeric value for all features except 'country'.")
    input_data[feature] = value

# Label Encoding for 'country'
label = LabelEncoder()
input_data['country'] = label.fit_transform([input_data['country']])

# Convert the dictionary to a NumPy array
input_data_as_np_array = np.array([list(input_data.values())])

# Input data Standardization
scaler = StandardScaler()
std_input_data = scaler.fit_transform(input_data_as_np_array)

# Make a prediction using the model
prediction = model.predict(std_input_data)
print("prediction: ",prediction)
```
